```python
import pandas as pd

reactores = pd.read_csv('/content/IAEA_Reactores_Nucleares_Investigacion.csv')
```

```python
# Paises
paises = reactores['Pais']
paises = paises.fillna('No registrado')
paises = paises.unique()
paises = pd.DataFrame(paises, columns=['nombre'])
paises.to_csv('paises.csv', sep=';', index=False)
paises
```

```python
# Ciudades

ciudades = reactores[['Pais', 'Ciudad']]
ciudades['Pais'] = ciudades['Pais'].apply(lambda x: list((paises[paises['nombre'] == x].index))[0] + 1)
ciudades = ciudades.fillna('No registrado')
ciudades = ciudades.drop_duplicates(subset='Ciudad')
ciudades.columns = ['pais_id', 'nombre']
ciudades = ciudades.reset_index()
ciudades = ciudades[['pais_id', 'nombre']]
ciudades.to_csv('ciudades.csv', sep=';', index=False)
ciudades
```

```python
# Tipo de reactores

tipo_reactores = reactores['Tipo'].unique()
tipo_reactores = pd.DataFrame(tipo_reactores, columns=['nombre'])
tipo_reactores = tipo_reactores.fillna('No registrado')
ciudades = ciudades.drop_duplicates(subset='nombre')
tipo_reactores.to_csv('tipo_reactores.csv', sep=';', index=False)
tipo_reactores
```

```python
# Estados reactores

estados_reactores = reactores['Estado'].unique()
estados_reactores = pd.DataFrame(estados_reactores, columns=['nombre'])
estados_reactores = estados_reactores.fillna('No registrado')
estados_reactores.to_csv('estados_reactores.csv', sep=';')
estados_reactores
```

```python
# Reactores

reactores['Ciudad'] = reactores['Ciudad'].fillna('No registrado')
reactores['Tipo'] = reactores['Tipo'].fillna('No registrado')
reactores['Estado'] = reactores['Estado'].fillna('No registrado')
reactores['Ciudad'] = reactores['Ciudad'].apply(lambda x: list((ciudades[ciudades['nombre'] == x].index))[0] + 1)
reactores.columns = ['nombre', 'Pais', 'ciudad_id', 'tipo_reactor_id', 'potencia_termica', 'estado_reactor_id', 'fecha_primera_reaccion']
reactores = reactores[['nombre', 'ciudad_id', 'tipo_reactor_id', 'potencia_termica', 'estado_reactor_id', 'fecha_primera_reaccion']]
reactores['estado_reactor_id'] = reactores['estado_reactor_id'].apply(lambda x: list((estados_reactores[estados_reactores['nombre'] == x].index))[0] + 1)
reactores['tipo_reactor_id'] = reactores['tipo_reactor_id'].apply(lambda x: list((tipo_reactores[tipo_reactores['nombre'] == x].index))[0] + 1)
reactores['fecha_primera_reaccion'] = pd.to_datetime(reactores['fecha_primera_reaccion'])
reactores.to_csv('reactores.csv', sep=';', index=False)
reactores
```
